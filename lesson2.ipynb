### Task2

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)

plt.axis([0.5, 9.5, -2.5, 2.5])
plt.plot(t, f, color='green')

plt.title(' f(t)', fontweight='bold')
plt.xlabel(' t')
plt.ylabel(' f')

plt.grid()
plt.show()
```
